```python
class createPlot():
    def __init__(self):
        self.info_dict = {}
        self.read_info()
    def read_info(self):
        path = "C:\\Users\\rando\\Desktop\\gas_prices.txt"
        with open(path, 'r') as file:
            header = file.readline()
            for line in file:
                if line.replace('\n','') != '':
                    self.add_info(line)
                #print(line)
    def add_info(self,line_str):
        line_arr = line_str.split(',')
        price = float(line_arr[2][line_arr[2].find('$')+1:line_arr[2].find('.')+3])
        for key in self.info_dict:
            if key == line_arr[1]:
                self.info_dict[key]['date_arr'].append(line_arr[0])
                self.info_dict[key]['price_arr'].append(price)
                return True
        new_key = line_arr[1]
        self.info_dict[new_key] = {}
        self.info_dict[new_key]['date_arr'] = []
        self.info_dict[new_key]['price_arr'] = []
        self.info_dict[new_key]['price_arr'].append(price)
        self.info_dict[new_key]['date_arr'].append(line_arr[0])
    def getSmallestPrice(self):
        smallest=100000
        biggest = 0
        for key in self.info_dict:
            for price in self.info_dict[key]['price_arr']:
                if price < smallest:
                    smallest = price
                if price > biggest:
                    biggest = price
        return (smallest,biggest)
    def plot(self):
        plt.xlabel('Date Range')
        plt.ylabel('Price ($)')
        smallest = self.getSmallestPrice()[0]
        biggest = self.getSmallestPrice()[1]
        plt.ylim(ymin=smallest-.01,ymax=biggest+.01)
        for key in self.info_dict:
            plt.plot(self.info_dict[key]['date_arr'],self.info_dict[key]['price_arr'],label=key)
        plt.legend()
```

```python
d = createPlot()
d.plot()
```
