```python
import sqlite3
connection = sqlite3.connect("northwind.db")
crsr = connection.cursor()

sql = "";
while sql != "stop":
    sql = prompt_questions()
    if sql != "stop":
        crsr.execute(sql)
        connection.commit()
connection.close() 

def prompt_questions():
    run = True;
    while(run):
        print("1. add a customer")
        print("2. add an order")
        print("3. remove an order")
        print("4. ship an order")
        print("5. print pending orders (not shipped yet) with customer information")
        print("6. restock products")
        print("7. exit")
        responce = input();
        if(not responce.isdigit()):
            print("please insert a valid entry")
        elif(responce == '1'):
            cust_id = input("What is the Customer ID?")
            comp_name = input("What is the Company Name?")
            contac_name = input("What is the Contact Name?")
            contac_tit = input("What is the Contact Title?")
            add = input("What is the address?")
            city = input("What is the city?")
            region = input("What is the region?")
            p_code = input("What is the Postal Code?")
            country = input("What is the Country?")
            phone = input("What is the phone number?")
            fax = input("What is the fax number?")
            customer_sql= """
            INSERT INTO Customers
            VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);
            """ % (cust_id,comp_name,contac_name,contac_tit,add,city,region,p_code,country,phone,fax,)
            return customer_sql
        elif(responce == '2'):
            order_id = input("What is the order ID?")
            cust_id = input("What is the Customer ID?")
            emp_id = input("What is the Employee ID?")
            order_date = input("What is the order date?")
            req_date = input("WHat is the required date?")
            ship_date = input("WHat is the shipped date?")
            ship_via = input("What is the ship via?")
            freight = input("What is the freight?")
            ship_name = input("What is the ship name?")
            ship_add = input("What is the ship address?")
            ship_city = input("What is the ship city?")
            ship_reg = input("What is the ship region?")
            ship_pcode = input("What is the ship postal code?")
            ship_cntry = input("What is the ship country?")
            order_sql = """
            INSERT INTO Orders
            VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);
            """ % (order_id,cust_id,emp_id,order_date,req_date,ship_date,ship_via,freight,ship_name,ship_add,ship_city,ship_reg,ship_pcode,ship_cntry,)
            return order_sql
        elif(responce == '3'):
            order_id = input("What is the order ID?")
            del_order_sql = """
            DELETE FROM Orders
            WHERE OrderID == '%s';
            DELETE FROM Order_details
            WHERE OrderID == '%s';
            """ % (order_id,order_id,)
            return del_order_sql
        elif(responce == '4'):
            order_id = input("What is the order ID?")
            date = input("What is the new ship date?")
            ship_sql = """
            UPDATE Orders
            SET ShippedDate = '%s'
            WHERE OrderID == '%s';
            """ % (date,order_id,)
            return ship_sql
        elif(responce == '5'):
            pending_orders_sql = """
            SELECT * FROM Customers
            INNER JOIN Products
            WHERE Products.ReorderLevel == 'pending';
            """
            return pending_orders_sql
        elif(responce == '6'):
            restock_sql = """
            UPDATE Products
            SET UnitsInStock = 10;
            """
            return restock_sql
        elif(responce == '7'):
            run = False
            return "stop"
        else:
            print("please insert a valid entry")
```
