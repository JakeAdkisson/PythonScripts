```python
import urllib.request
import codecs
```

```python
req = urllib.request.Request('https://www.crunchyroll.com/dr-stone/episode-1-stone-world-786941')
response = urllib.request.urlopen(req)
the_page = response.read()
the_page = the_page.decode("utf8")
#print(the_page)
file = codecs.open("crunchyroll.html", "w", "utf-8")
#file.write(u'\ufeff')
file.write(the_page)
file.close()
```
